```python
def most_score_runs_against(batsman_name):
  bn = deliveries[deliveries['batsman']==batsman_name]
  return bn.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
most_score_runs_against('V Kohli')
```
